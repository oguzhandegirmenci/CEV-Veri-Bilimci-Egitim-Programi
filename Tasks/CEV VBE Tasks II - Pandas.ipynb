**TASK-4B:** Import the datasets for the USA population example.

Then you are supposed to change one of the names of the states in areas by hand to a capital one such as from Alabama to ALABAMA.

Then you should change all of the state names in areas and abbrevs to uppercase letters.

Then you should add another column to the area data frame as km square and convert the values in the area column to km square in the new column.

Please merge the pandas’ data frames as done in the presented example and please name this dataframe as "final_df".

Then change 3 non-null values from this dataframe to null randomly for each of the following columns: state/region, state, year, ages.

Then please find a practical way to fill in these nan values in a systematic way.


```python
df_pop = pd.read_csv("state-population.csv")

df_pop.head()
```

```python
idx = ((df_pop["ages"] == "total") & (df_pop["year"] == 2012) & (df_pop["state/region"] != "USA") & (df_pop["state/region"] != "PR"))

df_pop_2 = df_pop[idx].reset_index().drop(["index", "ages", "year"], axis=1)

df_pop_2.head()
```

```python
df_states = pd.read_csv("state-abbrevs.csv")

dict_states = df_states.set_index("abbreviation")["state"].to_dict()
dict_states_upper = {k.upper():v.upper() for k, v in dict_states.items()}

df_pop_2["state/region"] = df_pop_2["state/region"].replace(dict_states_upper)
df_pop_2 = df_pop_2.rename(columns = {"state/region": "state"})

df_pop_2.head()
```

```python
df_area = pd.read_csv("state-areas.csv")

df_area["area"] = round(df_area["area (sq. mi)"] * 2.589988) # sq mi to sq km
df_area = df_area.drop(["area (sq. mi)"], axis=1)
df_area["state"] = df_area["state"].str.upper()

df_area.head()
```

```python
final_df = pd.merge(df_pop_2, df_area)
final_df.head()
```

```python
final_df.loc[1, "state"] = np.NaN
final_df.loc[2, "population"] = np.NaN
final_df.loc[3, "area"] = np.NaN
final_df.head()
```

```python
final_df["state"].fillna("UNDEFINED", inplace=True)
final_df.fillna(final_df.mean(), axis=0, inplace=True)
final_df.head()
```

**TASK-5:** Please use the dataset that has been shared during the online coding session. Then do all of the following operations especially including 
- aggregate
- filter
- transform
- apply


```python
df_5 = sns.load_dataset("planets")
df_5.head()
```

```python
df_5.dropna().groupby("year")["mass"].agg([min, np.median, max, np.sum]).head()
```

```python
def filter_func(x):
    return x["mass"].std() > 4

df_5.dropna().groupby("year").filter(filter_func)
```

```python
df_5.groupby("year").transform(lambda x: x - x.mean())
```

```python
def norm_by_data2(x):
    x["mass"] /= x["distance"].sum()
    return x

df_5.groupby("year").apply(norm_by_data2)
```
