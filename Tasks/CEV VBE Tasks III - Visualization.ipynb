**Task-1D:** You are asked to replicate the example given for continuous errors with the Gaussian process and please run it with 5 different parameters and summarize the differences between these.

**Task-3A:** Please make an annual figure as in 04.09 but with random numbers that will mainly include the months of a year and you are responsible for giving text messages for all of the holidays in Turkey.


```python
plt.style.use('seaborn-whitegrid')
```

```python
# Noised data
np.random.seed(137)
t = np.linspace(0, 10, 365)
p = (np.sin(t)/5+2)*2000
percentage = 0.5
n = np.random.normal(0, p.std(), t.size) * percentage
pn = p + n
plt.plot(t, pn);
```

```python
index = pd.date_range('2021-01-01', '2021-12-31')
data = pn

births = pd.DataFrame({'year':index.year, 'month':index.month, 'day':index.day, 'births':data}, index=index)
```

```python
quartiles = np.percentile(births['births'], [25, 50, 75])
mu, sig = quartiles[1], 0.74 * (quartiles[2] - quartiles[0])
births = births.query('(births > @mu - 5 * @sig) & (births < @mu + 5 * @sig)')

births['day'] = births['day'].astype(int)

births.index = pd.to_datetime(10000 * births.year +
                              100 * births.month +
                              births.day, format='%Y%m%d')
births_by_date = births.pivot_table('births',
                                    [births.index.month, births.index.day])
births_by_date.index = [pd.datetime(2021, month, day)
                        for (month, day) in births_by_date.index]
```

```python
fig, ax = plt.subplots(figsize=(12, 4))
births_by_date.plot(ax=ax)

# Add labels to the plot
style = dict(size=10, color='red')

ax.text('2021-1-1', 3800, "Yılbaşı", **style)
ax.text('2021-4-23', 3700, "Ulusal Egemenlik \nve Çocuk Bayramı", ha='center', **style)
ax.text('2021-5-1', 4200, "Emek ve \nDayanışma Günü", ha='center', **style)
ax.text('2021-5-19', 3400, "Atatürk’ü Anma, Gençlik \nve Spor Bayramı", ha='center', **style)
ax.text('2021-7-15', 4100, "Demokrasi Bayramı", ha='center', **style)
ax.text('2021-8-30', 3800, "Zafer Bayramı ", ha='center', **style)
ax.text('2021-10-29', 3850, "Cumhuriyet Bayramı ", ha='center', **style);

# Format the x axis with centered month labels
ax.xaxis.set_major_locator(mpl.dates.MonthLocator())
ax.xaxis.set_minor_locator(mpl.dates.MonthLocator(bymonthday=15))
ax.xaxis.set_major_formatter(plt.NullFormatter())
ax.xaxis.set_minor_formatter(mpl.dates.DateFormatter('%h'));
```

```python
def format_func(value, tick_number):
    # find number of multiples of pi/2
    N = int(np.round(4 * value / np.pi))
    if N == 0:
        return "0"
    elif N == 1:
        return r"$\pi/4$"
    elif N == 2:
        return r"$\pi/2$"
    elif N == 3:
        return r"$3\pi/4$"
    elif N == 4:
        return r"$\pi$"
    elif N == 8:
        return r"$2\pi$"
    elif N == 10:
        return r"$5\pi/2$"
    elif N % 2 > 0:
        return r"${0}\pi/4$".format(N)
    elif N % 3 > 0:
        return r"${0}\pi/4$".format(N)
    elif N % 4 > 0:
        return r"${0}\pi/2$".format(N // 2)

ax.xaxis.set_major_formatter(plt.FuncFormatter(format_func))
fig
```
